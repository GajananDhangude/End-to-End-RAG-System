```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = text_splitter.split_documents(docs)
```

```python
pip install -qU langchain-huggingface
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name = "sentence-transformers/all-MiniLM-L6-v2")
```

```python
pip install -qU langchain-chroma
```

```python
from langchain_chroma import Chroma
vector_store = Chroma(
    collection_name = "attention-is-all-you-need",
    embedding_function = embeddings,
    persist_directory = "attention-is-all-you-need"
)
```

```python
vector_store.add_documents(chunks)
```

```python
similarity_search = vector_store.similarity_search(
    query = "What is Attention machanism ?", k = 2)
```

```python
for doc in similarity_search:
  print(doc.page_content[:400])
  print(doc.metadata)
```

```python
from langchain_community.retrievers import KNNRetriever
```

```python
retriever = KNNRetriever.from_documents(
    documents = chunks,
    embeddings = embeddings,
    top_k_retriever = 2
)
```

```python
docs = retriever.invoke(
    "What is attention mechanism?"
)

for d in docs:
    print("----")
    print(d.page_content[:300])
    print(d.metadata)
```

```python
import getpass
import os

if "GROQ_API_KEY" not in os.environ:
    os.environ["GROQ_API_KEY"] = getpass.getpass("Enter your Groq API key: ")
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model_name = "llama-3.3-70b-versatile",
    temperature=0.5
)
```

```python
messages = [
    (
        "system",
        "You are a helpful assistant that translates English to HIndi. Translate the user sentence.",
    ),
    ("human", "I love programming."),
]
ai_msg = llm.invoke(messages)
ai_msg
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_template(
   """
You are an AI assistant.

Using ONLY the information from the context below,
synthesize a clear and concise answer to the question.
Do NOT copy sentences verbatim from the context.
Explain in your own words.

If the answer is not present in the context, say:
"I don't know based on the provided documents."

Context:
{context}

Question:
{question}

Answer:
"""
)
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

```python
rag_chain = (
    {
        "context":retriever,
        "question":RunnablePassthrough()
    }
    | prompt
    | llm
    | StrOutputParser()
)
```

```python
responce  = rag_chain.invoke(
    "What is Attention mechanism?"
)

print(responce)
```
